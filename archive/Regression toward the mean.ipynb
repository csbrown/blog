```python
import numpy as np
import matplotlib.pyplot as plt

MEAN = 0
TRUE_STD = 1
def do_the_experiment(N):
    x_true = np.random.normal(MEAN,TRUE_STD,N)
    x_random = np.random.normal(0,1,N)
    x_random2 = np.random.normal(0,1,N)
    data1 = x_true + x_random
    data2 = x_true + x_random2
    order1 = sorted(range(N), key = lambda x: data1[x])
    order2 = sorted(range(N), key = lambda x: data2[x])
    regress_data = []
    for i in range(N):
        #this value quantifies regression toward the mean. 
        #  How far away from the mean at time 1, minus how far away from mean at time 2.
        #  The abs makes it so that "toward" is positive from either side
        regression_toward_the_mean = abs(data2[order1[i]]-MEAN) - abs(data1[order1[i]]-MEAN)
        regress_data.append(regression_toward_the_mean)
    print "lowest performers".ljust(37," "), np.mean(regress_data[0:N/5])
    print "highest performers".ljust(37," "), np.mean(regress_data[-N/5:])
    print "middling performers".ljust(37," "), np.mean(regress_data[N/2-N/5:N/2+N/5])
    print "average regression toward the mean".ljust(37," "), np.mean(regress_data)
    plt.plot(regress_data)
    plt.show()
```

```python
do_the_experiment(100)
```

```python
do_the_experiment(1000)
```
